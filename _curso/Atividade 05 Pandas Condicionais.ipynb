### Criando novas colunas

```python
df['Nascimento2'] = pd.to_datetime(df['Nascimento'])
```

```python
df['Idade'] = (pd.to_datetime('today') - df['Nascimento2']).dt.days
```

```python
df.head()
```

### Condicionais

```python
masculino = (df['Sexo'] == 'M')
```

```python
masculino
```

```python
velho = (df['Idade'] >= 21*365)
```

```python
velho
```

```python
df[masculino]
```

```python
df[velho]
```

```python
df[~masculino]
```

```python
df[~velho]
```

```python
df[masculino & velho]
```

```python
df[masculino | velho]
```

```python
df[~masculino & ~velho]
```

### Criando coluna com condicionais

```python
df['Target'] = 0
df.loc[masculino, 'Target'] = 1
```
